```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
star_data['Star color Type']=label.fit_transform(star_data['Star color'])
```

```python
star_data['Spectral Class Type']=label.fit_transform(star_data['Spectral Class'])
```

### **0: Red Dwarf, 1: Brown Dwarf, 2: White Dwarf, 3: Main Sequence ,4:  SuperGiants, 5: HyperGiants**

```python
star_data
```

```python
star_data=star_data.drop('Star color',axis=1)
```

```python
star_data=star_data.drop('Spectral Class',axis=1)
```

```python
x=star_data.iloc[:,:-1]
y=star_data['Spectral Class Type']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,predict)
```

```python
accuracy_score(y_test,predict)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
predict=model.predict(x_test)
```

```python
accuracy_score(y_test,predict)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
predict=model.predict(x_test)
accuracy_score(y_test,predict)
```
